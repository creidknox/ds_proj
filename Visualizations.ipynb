```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('backward_all.csv')
```

```python
df
```

```python
variables = df.columns[3:7]
variables
for n,var in enumerate(variables):
    print(n)
    print(var)
```

```python
f, axarr = plt.subplots(4, 1, figsize=(10,20))

for n,var in enumerate(variables):
    for label, sub_df in df.groupby(['database']):
        sub_df = sub_df.groupby('weeks_to_death').mean()
        sub_df = sub_df.rolling(window=5, min_periods=1).mean().reset_index()
        sub_df.plot(ax=axarr[n], label=label, x='weeks_to_death', y=var)
    axarr[n].set_title(var)
    axarr[n].invert_xaxis()

plt.show()
```
